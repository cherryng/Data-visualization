```python
#Input file from Pierre, with FRB, Date, and Integration on separate lines.
#Read them in to array and re-write to a new files (DataBase_NenufarTF_v7.txt) with the info on the same line.
#Then outside of python, I used egrep to remove the sources not relevant (e.g. Pulsars) to make DataBase_NenufarTF_v8.txt

with open('DataBase_NenufarTF_v6.txt', 'r') as file:
    lines = file.read().splitlines()

FRBs = []
Dates = []
Tints = []
format = '%Y%m%d'
ii = 0 

print(len(lines))
for i in range(int(len(lines)/4)):
    FRBs.append(str(lines[i*4]))
    #Date = (datetime.datetime.strptime(str(lines[i*4+1]),format))
    #Dates.append(Date)
    Dates.append(int(lines[i*4+1]))
    Tints.append(int(lines[i*4+3]))

with open('DataBase_NenufarTF_v7.txt', 'a') as f:
    for i in range(len(FRBs)):
        print(str(FRBs[i]),str(Dates[i]),str(Tints[i]))
        f.write("{0},{1},{2}\n".format(FRBs[i],Dates[i],Tints[i]))
f.close()
    
```

```python
import pylab as plt
import matplotlib as mpl
%matplotlib inline
import datetime
import numpy as np

FRBs = []
Dates = []
Tints = []
format = '%Y%m%d'

#Re-formatted input file, with FRB,Date,Integration time on each line.

with open('DataBase_NenufarTF_v8.txt', 'r') as data:

    for line in data:
        p = line.split(',')
        FRBs.append(p[0])
        Date = datetime.datetime.strptime(p[1],format)
        Dates.append(Date)
        Tints.append(int(p[2]))
 
    
print(FRBs[0:5])
print(Dates[0:5],len(Dates))

mpl.rcParams.update({'font.size': 18,'font.family': 'serif'})
f, (ax,ax2) = plt.subplots(1,2, gridspec_kw={'width_ratios':[3,1]}, figsize=(10,8), sharey=True)
plt.subplots_adjust(wspace=0.05)

#Plot obs cadence per source
ax.scatter(Dates,FRBs)
for tick in ax.get_xticklabels():
    tick.set_rotation(45) 
ax.grid()
ax.set_xlabel("Observation Date")

#Count the number of hours per source
Uniqs = np.unique(FRBs)
Counts =  np.zeros(len(Uniqs))
for i in range(len(Uniqs)):
    for j in range(len(FRBs)):
        if FRBs[j] == Uniqs[i]:
            Counts[i] = Tints[j]+Counts[i]
            
print(Uniqs) 
print(Counts)

hbars = ax2.barh(Uniqs,Counts/3600.)
ax2.grid()
ax2.grid(b=True, which='major', color='#ccc',linestyle='--')
ax2.bar_label(hbars, fmt='%.2f')
ax2.set_xlabel("Hours")
ax2.xaxis.set_major_locator(plt.MaxNLocator(3))

#Sum up total hours for all observations
total = 0
for i in range(len(Counts)):
    total = total + Counts[i]
print("Total time", np.round(total/3600.,2), "hr")
```
